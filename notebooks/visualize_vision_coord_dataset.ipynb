```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import random

def run_micro_visual_audit():
    data_dir = Path("../outputs/vision_dataset/2026_06_01/09_15_29/")
    images = np.load(data_dir / "images.npy")
    coords = np.load(data_dir / "coords.npy") # shape [N, 12]
    
    num_samples = len(images)
    sampled_indices = random.sample(range(num_samples), 16)
    
    fig, axs = plt.subplots(4, 4, figsize=(12, 12))
    fig.suptitle("Perfect Spatial Alignment Audit (-0.5 Pixel Compensation)", fontsize=16)
    
    for idx, ax in zip(sampled_indices, axs.flatten()):
        img = images[idx]
        coord = coords[idx]
        
        ax.imshow(img)
        
        entities_coords = {
            "Prey": (coord[0], coord[1], "cyan"),       
            "Pred": (coord[2], coord[3], "magenta"),    
            "Shelter": (coord[4], coord[5], "yellow"),  
            "Land1": (coord[6], coord[7], "orange"),    
            "Land2": (coord[8], coord[9], "orange"),
            "Land3": (coord[10], coord[11], "orange")
        }
        
        for name, (cx, cy, color) in entities_coords.items():
            # Half-pixel offset compensation
            px = (cx + 1.0) / 2.0 * 128.0 - 0.5
            py = (cy + 1.0) / 2.0 * 128.0 - 0.5

            ax.plot(px, py, color=color, marker='+', markersize=12, markeredgewidth=1.5)
            
        ax.set_title(f"ID: {idx}", fontsize=9)
        ax.axis('off')
        
    plt.tight_layout()
    plt.show()

run_micro_visual_audit()
```
